Analyze Zomato restaurant data to uncover:

- Top and bottom performing restaurants


## 1. 📦 Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Plot styling
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
```

## 2. 📂 Load Dataset

```python
# Load the Zomato dataset
# Update the path below to point to your local CSV file
df = pd.read_csv('Zomato data .csv')

print(f"Dataset shape: {df.shape}")
df.head(10)
```

## 3. 🔍 Data Exploration

```python
# Dataset structure and data types
df.info()
```

```python
# Check for missing values
missing = df.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0])
```

```python
# Statistical summary of numeric columns
df.describe()
```

## 4. 🧹 Data Cleaning

```python
# The 'rate' column contains values like '4.1/5' — extract the numeric part
def parse_rate(value):
    """Convert rate string '4.1/5' to float 4.1"""
    try:
        return float(str(value).split('/')[0])
    except ValueError:
        return np.nan

df['rate'] = df['rate'].apply(parse_rate)

print("Rate column cleaned. Sample values:")
print(df['rate'].head())
```

## 5. 📊 Analysis & Visualizations

### 5.1 Distribution of Restaurant Types

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.countplot(x=df['listed_in(type)'], hue=df['listed_in(type)'], ax=ax, legend=False)
ax.set_title('Number of Restaurants by Type')
ax.set_xlabel('Restaurant Type')
ax.set_ylabel('Count')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

> **Insight:** The majority of restaurants are listed under the **Dining** category, making it the most common restaurant type on Zomato.

### 5.2 Total Votes by Restaurant Type

```python
votes_by_type = df.groupby('listed_in(type)')['votes'].sum().reset_index()

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(votes_by_type['listed_in(type)'], votes_by_type['votes'], color='steelblue', marker='o', linewidth=2)
ax.set_title('Total Votes by Restaurant Type')
ax.set_xlabel('Restaurant Type')
ax.set_ylabel('Total Votes')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

> **Insight:** **Dining restaurants** receive the highest number of total votes, indicating strong customer engagement with sit-down dining experiences.

### 5.3 Rating Distribution

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(df['rate'].dropna(), bins=15, color='steelblue', edgecolor='white')
ax.set_title('Distribution of Restaurant Ratings')
ax.set_xlabel('Rating')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

> **Insight:** Most restaurants are rated between **3.5 and 4.0**, suggesting generally positive customer satisfaction across the platform.

### 5.4 Online vs Offline Orders — Impact on Ratings

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.boxplot(x='online_order', y='rate', data=df, ax=ax, palette='Set2')
ax.set_title('Ratings: Online vs Offline Orders')
ax.set_xlabel('Online Order Available')
ax.set_ylabel('Rating')
plt.tight_layout()
plt.show()
```

> **Insight:** Restaurants that accept **online orders** tend to receive higher ratings, suggesting online availability correlates with better customer experience.

### 5.5 Order Mode Heatmap by Restaurant Type

```python
pivot = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)

fig, ax = plt.subplots(figsize=(8, 5))
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlOrRd', ax=ax)
ax.set_title('Order Mode Distribution by Restaurant Type')
ax.set_xlabel('Online Order')
ax.set_ylabel('Restaurant Type')
plt.tight_layout()
plt.show()
```

> **Insight:** **Dining restaurants** predominantly receive offline orders, while **Cafes** are the top category for online orders — a key difference in customer behavior.

### 5.6 Top 10 & Bottom 10 Restaurants by Rating

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

top10 = df.nlargest(10, 'rate')
axes[0].barh(top10['name'], top10['rate'], color='#4CAF50')
axes[0].set_title('Top 10 Restaurants by Rating')
axes[0].set_xlabel('Rating')
axes[0].invert_yaxis()

bottom10 = df.nsmallest(10, 'rate')
axes[1].barh(bottom10['name'], bottom10['rate'], color='#E91E63')
axes[1].set_title('Bottom 10 Restaurants by Rating')
axes[1].set_xlabel('Rating')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

> **Insight:** **Onesta** ranks highest among rated restaurants, while **Nandhini Deluxe** has the lowest rating in the dataset.

### 5.7 Table Booking Behavior

```python
fig, ax = plt.subplots(figsize=(6, 5))
sns.countplot(x=df['book_table'], hue=df['book_table'], ax=ax, palette='pastel', legend=False)
ax.set_title('Table Booking Frequency')
ax.set_xlabel('Table Booked')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

> **Insight:** The vast majority of customers **do not book tables in advance**, suggesting most dining is walk-in or spontaneous.

### 5.8 Cost Distribution by Restaurant Type

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['listed_in(type)'], df['approx_cost(for two people)'], alpha=0.5, color='coral')
ax.set_title('Approximate Cost for Two by Restaurant Type')
ax.set_xlabel('Restaurant Type')
ax.set_ylabel('Approx Cost (₹) for Two People')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

**Insight — Spending Patterns by Category:**
| Category | Approx Cost Range (₹ for 2) |
|----------|------------------------------|
| Buffet   | ₹250 – ₹800                 |
| Cafes    | ₹190 – ₹800+                |
| Dining   | ₹10 – ₹850+                 |
| Other    | ₹400 – ₹800                 |
